### **Content Based Recommendation System**

```python
movies_cleaned.head(1)['overview']
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfv = TfidfVectorizer(min_df=3,  max_features=None,
            strip_accents='unicode', analyzer='word',token_pattern=r'\w{1,}',
            ngram_range=(1, 3),
            stop_words = 'english')

# Filling NaNs with empty string
movies_cleaned['overview']=movies_cleaned['overview'].fillna('')
```

```python
# Fitting the TF-IDF on the 'overview' text
tfv_matrix=tfv.fit_transform(movies_cleaned['overview'])
```

```python
tfv_matrix
```

```python
tfv_matrix.shape
```

```python
from sklearn.metrics.pairwise import sigmoid_kernel
sig=sigmoid_kernel(tfv_matrix,tfv_matrix)
```

```python
sig.shape
```

```python
sig[0]
```

```python
indices=pd.Series(movies_cleaned.index,index=movies_cleaned['original_title']).drop_duplicates()
indices
```

```python
indices['Harry Potter and the Half-Blood Prince']
```

```python
list(enumerate(sig[indices['Harry Potter and the Half-Blood Prince']]))
```

```python
sorted(list(enumerate(sig[indices['Harry Potter and the Half-Blood Prince']])),key=lambda x: x[1],reverse=True)
```

```python
def give_rec(title,sig=sig):
  # Get the index corresponding to original_title
  idx=indices[title]

  # Get the pairwsie similarity scores
  sig_scores=list(enumerate(sig[idx]))

  # Sort the movies
  sorted_sig_scores=sorted(sig_scores,key=lambda x:x[1], reverse = True)

  # Scores of the 10 most similar movies
  sorted_sig_scores = sorted_sig_scores[1:11]

  # Movie indices
  movie_indices= [i[0] for i in sorted_sig_scores]

  # Top 10 most similar movies
  return movies_cleaned['original_title'].iloc[movie_indices]
```

```python
give_rec('Avatar')
```
